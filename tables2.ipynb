```python
import mysql.connector

from sqlalchemy import create_engine

# create sqlalchemy engine
engine = create_engine("mysql+pymysql://{user}:{pw}@localhost/{db}"
                       .format(user="jbatista",
                               pw="",
                               db="final_schema"))

```

```python
cnx = mysql.connector.connect(user = 'jbatista',
                                password = '',
                                host = '127.0.0.1')
```

```python
cnx.is_connected()
```

```python
cursor = cnx.cursor()
cursor
```

```python
query_courses = '''SELECT * 
FROM project_data_lake.courses;'''

query_courses
```

```python
cursor.execute(query_courses)
```

```python
courses_result = cursor.fetchall()
courses_result
```

```python
import pandas as pd

courses_df = pd.DataFrame(courses_result)
courses_df
```

```python
column_names = [head[0] for head in cursor.description]
```

```python
courses_df.columns = column_names
courses_df
```

```python
courses_df.drop(["index"], axis=1, inplace=True)
courses_df
```

```python
courses_df.reset_index(inplace=True)
courses_df
```

```python
courses_df.to_sql('courses', con = engine, if_exists = 'replace', chunksize = 1000)
```
