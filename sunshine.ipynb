```python
import pandas as pd

europe = pd.read_html('https://en.wikipedia.org/wiki/List_of_cities_by_sunshine_duration')[4]
europe
```

```python
europe_df = europe[['City', 'Year']]
condition = europe_df['City'].isin(['Amsterdam', 'Lisbon', 'Berlin', 'Paris', 'Madrid', 'Barcelona'])
europe_df = europe_df[condition]
```

```python
north_america = pd.read_html('https://en.wikipedia.org/wiki/List_of_cities_by_sunshine_duration')[5]
north_america
```

```python
north_american_df = north_america[['City', 'Year']]
condition = north_american_df['City'].isin(['San Francisco', 'New York City', 'Mexico City'])
north_american_df = north_american_df[condition]
```

```python
south_america = pd.read_html('https://en.wikipedia.org/wiki/List_of_cities_by_sunshine_duration')[6]
south_america
```

```python
south_american_df = south_america[['City', 'Year']]
condition = south_american_df['City'].isin(['São Paulo'])
south_american_df = south_american_df[condition]
```

```python
sunshine = pd.concat([south_american_df, north_american_df, europe_df])
sunshine
```

```python
sunshine.reset_index(names="city_id", inplace=True)
```

```python
sunshine

sunshine["location_id"] = [16088, 16022, 16328, 15704, 16109, 15901, 16086, 16709, 16377, 16376]
sunshine
```
